# Preprocesamiento de datos

Se instan los diferentes formas de preprocesamiento que se usarán en las pruebas

Funcion para limpiar el texto de las canciones y de los diferentes datasets.

```python
# Funcion para limpiar texto
def clean_text(text):
    text = emoji.demojize(text)
    text = text.lower()
    text = re.sub(r'(?<=@)\w+','',text)
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "can not ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"'m", " am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text)
    text = re.sub("[^a-zA-Z0-9 ]", "", text)
    text = re.sub('verse', "", text)
    text = re.sub('chorus', '', text)
    text = re.sub('instrumental', '', text)
    text = re.sub('guitar solo', '', text)
    text = re.sub('solo guitar', '', text)
    text = re.sub('intro', '', text)
    text = re.sub('user', '', text)
    text = re.sub('hashtag', '', text)
    text = re.sub('url', '', text)
    text = re.sub('oh', '', text)
    text = re.sub('ooh', '', text)
    text = re.sub('na', '', text)
    text = re.sub('nana', '', text)
    text = re.sub('eh', '', text)
    text = re.sub('da', '', text)
    text = text.strip(' ')
    final_text=""

    return text
```

Función para eliminar las palabras vacias. Se lematizan las palabras para que la deteccion de las palabras vacias sea mas eficaz.

```python
def deleteStopwords(text):
  textFiltered =""
  words = tokenizer.tokenize(text)
  for w in words:
    if w not in english_stop:
        textFiltered=textFiltered+ " " + lemmatizer.lemmatize(w)

  return textFiltered
```

# Evaluador de emociones

Dado un texto se analizará y se devolvera una puntuacion normalizada de cada emoción

```python
def getIntensityLexicon(path):
  lexicon = dict()
  f = open(path)
  for line in f.readlines():
    lexicon[line.split()[0]]=float(line.split()[1])
  f.close()
  return lexicon
```

```python
print(path)
intensityWords=[]
f = open(path+"Lexicones/NRC-Emotion-Intensity/anger-NRC-EmoIntv1-withZeroIntensityEntries.txt")
for line in f.readlines():
  intensityWords.append(line.split()[0])
f.close()
print(path)
path_lexicon = path+"Lexicones/NRC-Emotion-Intensity/"
print(path_lexicon)
anger_intensity_lexicon = getIntensityLexicon(path_lexicon + "anger-NRC-EmoIntv1-withZeroIntensityEntries.txt")
disgust_intensity_lexicon = getIntensityLexicon(path_lexicon + "disgust-NRC-EmoIntv1-withZeroIntensityEntries.txt")
fear_intensity_lexicon = getIntensityLexicon(path_lexicon + "fear-NRC-EmoIntv1-withZeroIntensityEntries.txt")
joy_intensity_lexicon = getIntensityLexicon(path_lexicon + "joy-NRC-EmoIntv1-withZeroIntensityEntries.txt")
sadness_intensity_lexicon = getIntensityLexicon(path_lexicon + "sadness-NRC-EmoIntv1-withZeroIntensityEntries.txt")
surprise_intensity_lexicon = getIntensityLexicon(path_lexicon + "surprise-NRC-EmoIntv1-withZeroIntensityEntries.txt")
trust_intensity_lexicon = getIntensityLexicon(path_lexicon + "trust-NRC-EmoIntv1-withZeroIntensityEntries.txt")
anticipation_intensity_lexicon = getIntensityLexicon(path_lexicon + "anticipation-NRC-EmoIntv1-withZeroIntensityEntries.txt")

def findOnIntentityLexicon(word):
  findedOn = []
  for i in range(8):
    findedOn.append(0.0)

  if word in intensityWords:
    findedOn[0]=anger_intensity_lexicon[word]
    findedOn[1]=disgust_intensity_lexicon[word]
    findedOn[2]=fear_intensity_lexicon[word]
    findedOn[3]=joy_intensity_lexicon[word]
    findedOn[4]=sadness_intensity_lexicon[word]
    findedOn[5]=surprise_intensity_lexicon[word]
    findedOn[6]=trust_intensity_lexicon[word]
    findedOn[7]=anticipation_intensity_lexicon[word]


  return findedOn
```

```python
def songScorer(text,spliter):
  # Evaluacion final que sera devuelta
  findedWords = 0
  total = 0
  scores=[0,0,0,0,0,0,0,0]
  no_emotion = scores

  for verse in text.split(spliter):
    doc = nlp(verse)
    for token in doc:
      negated = False
      emotion = findOnIntentityLexicon(token.text)
      if emotion != no_emotion:
        # Tratamiento de la negacion
        for child in token.head.children:
          if child.dep_ == "neg":
            negated = True

        if negated == False:
          for i in range(8):
            total = total + emotion[i]
            scores[i] = scores[i] + emotion[i]

  if total != 0:
    for i in range(8):
      scores[i] = scores[i]/total

  return scores
```

```python
def emotionAnalysis(emocion,nombreCorpus,datos,etiquetas,spliter):
  datos_emocion=dict()
  categories = ["joy","trust","fear","surprise","sadness","disgust","anger","anticipation"]
  indice_emocion = categories.index(emocion)
  emotionData=[]
  indices=[]
  relacionEmociones = [0,0,0,0,0,0,0,0]
  for i in range(0,len(etiquetas)):
    if etiquetas[i][indice_emocion]==1:
      emotionData.append(datos[i])
      indices.append(i)
      for j in range(0,8):
        if j != indice_emocion:
          relacionEmociones[j]=relacionEmociones[j]+etiquetas[i][j]

  datos_emocion["indices"]=indices
  datos_emocion["relaciones"]=relacionEmociones
  # x-coordinates of left sides of bars
  left = [1,2,3,4,5,6,7,8]

  # heights of bars
  height = relacionEmociones

  # labels for bars
  tick_label = categories

  # plotting a bar chart
  plt.bar(left, height, tick_label = tick_label, width = 0.8, color = ['red', 'green', 'blue','orange'])

  # naming the x-axis
  plt.xlabel('Emociones')
  # naming the y-axis
  plt.ylabel('numero de apariciones')
  # plot title
  titulo = "Emociones relacionadas con " + emocion + " ("+nombreCorpus+")"
  plt.title(titulo)
  # function to show the plot
  plt.show()

  print("")
  songs_scores=[0,0,0,0,0,0,0,0]
  text=""
  for i in emotionData:
    score = songScorer(i,spliter)
    text = text + " " + deleteStopwords(i)
    for j in range(0,8):
      songs_scores[j]=songs_scores[j]+score[j]

  for i in range(0,8):
    songs_scores[i]=songs_scores[i]/len(emotionData)

  datos_emocion["intensidad_emociones_relacionadas"]=songs_scores
  # x-coordinates of left sides of bars
  left = [1,2,3,4,5,6,7,8]

  # heights of bars
  height = songs_scores

  # labels for bars
  tick_label = categories

  # plotting a bar chart
  plt.bar(left, height, tick_label = tick_label, width = 0.8, color = ['red', 'green', 'blue','orange'])

  # naming the x-axis
  plt.xlabel('Emociones')
  # naming the y-axis
  plt.ylabel('Intensidad media')
  # plot title
  titulo = "Intensidad media de las emociones relacionadas con " + emocion + " ("+nombreCorpus+")"
  plt.title(titulo)
  # function to show the plot
  plt.show()
  print("")

  text = re.sub("/",'',text)
  tokens=tokenizer.tokenize(text)
  text = Text(tokens)
  fdist = FreqDist(text)
  most_common = fdist.most_common(30)
  print("Palabras identificativas de las canciones con etiqueta " + emocion + ":")
  print(most_common)

  datos_emocion["palabras_identificativas_canciones"]=most_common

  lexicon=[]
  if emocion==categories[0]:
    lexicon=joy_binary_lexicon
  elif emocion==categories[1]:
    lexicon=trust_binary_lexicon
  elif emocion==categories[2]:
    lexicon=fear_binary_lexicon
  elif emocion==categories[3]:
    lexicon=surprise_binary_lexicon
  elif emocion==categories[4]:
    lexicon=sadness_binary_lexicon
  elif emocion==categories[5]:
    lexicon=disgust_binary_lexicon
  elif emocion==categories[6]:
    lexicon=anger_binary_lexicon
  elif emocion==categories[7]:
    lexicon=anticipation_binary_lexicon

  palabras_emocion = dict()
  for i in tokens:
    if i in lexicon:
      if i not in palabras_emocion:
        palabras_emocion[i]=1
      else:
        palabras_emocion[i]=palabras_emocion[i]+1

  palabras_emocion = sorted(palabras_emocion.items(), key=lambda x:x[1],reverse=True)
  lista_palabras=list(palabras_emocion)

  datos_emocion["palabras_identificativas_emocion"]=lista_palabras

  print("")
  print("Palabras identificativas de la emocion " + emocion + " en las canciones:")
  for i in range(0,10):
    print(lista_palabras[i])
  return datos_emocion
```

#Undersampling

Dado que los corpus no están equilibrados es necesario realizar undersampling para proporcionar a los algorimos subconjuntos de datos con iguales numero de datos positivos y negativos

```python
import random
def undersampling(datosEmocion, dataset_label):
  indices=[]
  if len(datosEmocion["indices"])*2 >= len(dataset_label):
    for i in range(0,len(dataset_label)):
      indices.append(i)
  else:
    for i in datosEmocion["indices"]:
      indices.append(i)

    while len(indices) < len(datosEmocion["indices"])*2:
      posible = random.randint(0, len(dataset_label)-1)

      if posible not in indices:
        indices.append(posible)

  return indices
```

```python
def resultados(titulo,pred,tags_test):
  categories = ["joy","trust","fear","surprise","sadness","disgust","anger","anticipation"]
  print("********* " + titulo + " *********")
  print("Accuracy Score: " + str(accuracy_score(pred, tags_test)))
  print("Coverage Error: " + str(coverage_error(pred, tags_test)))
  print("Label Ranking Average Precision Score: " + str(label_ranking_average_precision_score(pred, tags_test)))
  print("Label Ranking Loss Score: " + str(label_ranking_loss(pred, tags_test)))
  print("Classification Report:")
  print(classification_report(pred, tags_test, target_names=categories))
  print("Matrices de confusion de cada etiqueta ("+str(categories)+")")
  print(multilabel_confusion_matrix(pred, tags_test))
  print("")
  print("Accuraccy por emocion:")
  for i in range(0,8):
    aciertos=0
    for j in range(0,len(pred)):
      if pred[j][i]==tags_test[j][i]:
        aciertos=aciertos+1

    print("   - " + categories[i] + ": " + str(aciertos/len(pred)))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
import numpy as np
from statistics import mean
def test_models_EdmondsDance(model, indices, corpus_data, corpus_label, stopwords, features=None):
  categories = ["joy","trust","fear","surprise","sadness","disgust","anger","anticipation"]
  for i in range(0,8):
    print("\n*********************** " + categories[i] + " ***********************")
    data=[]
    label=[]
    features_emotion=[]
    for j in indices[i]:
      data.append(corpus_data[j])
      label.append(corpus_label[j][i])

    if features is not None:
      for j in indices[i]:
        features_emotion.append(features[j])

    if stopwords is True:
      data_stopwords=[]
      for j in data:
        data_stopwords.append(deleteStopwords(j))

      data = data_stopwords

    #X = vectorizer.fit_transform(data)

    #if features is not None:
      #X=hstack((X, features))

    accuracy_scores=[]
    f1_scores=[]
    precision_scores=[]
    recall_scores=[]

    skf = StratifiedKFold(n_splits=5)

    for train_index, test_index in skf.split(data, label):
      x_train_fold, x_test_fold = np.array(data)[train_index], np.array(data)[test_index]
      y_train_fold, y_test_fold = np.array(label)[train_index], np.array(label)[test_index]

      x_train = vectorizer.fit_transform(x_train_fold)
      x_test = vectorizer.transform(x_test_fold)

      if features is not None:
        features_train, features_test = np.array(features_emotion)[train_index], np.array(features_emotion)[test_index]
        sparce_train = csr_matrix(features_train)
        sparce_test = csr_matrix(features_test)
        x_train = hstack((x_train, sparce_train))
        x_test = hstack((x_test, sparce_test))

      model.fit(x_train, y_train_fold)
      prediction=model.predict(x_test)

      accuracy_scores.append(accuracy_score(y_test_fold, prediction))
      f1_scores.append(f1_score(y_test_fold, prediction))
      precision_scores.append(precision_score(y_test_fold, prediction))
      recall_scores.append((y_test_fold, prediction))

    print('Accuracy:', accuracy_scores)
    print('\t Accuracy Media:',
          mean(accuracy_scores)*100, '%')

    print('\n F1:', f1_scores)
    print('\t F1 Media:',
          mean(f1_scores)*100, '%')

    print('\nRecall:', recall_scores)
    print('\t Recall Media:',
          mean(recall_scores)*100, '%')

    print('\nPrecision:', precision_scores)
    print('\t Precision Media:',
          mean(precision_scores)*100, '%')
```

```python
labels_songs_corpus=[]
for i in range(0,8):
  data_train=[]
  label_train=[]
  scores_train=[]

  for j in indicesEdmondsDance[i]:
    data_train.append(no_stop_EdmondsDance_data[j])
    label_train.append(EdmondsDance_label[j][i])
    scores_train.append(scores_EdmondsDance[j])

  X=vectorizer.fit_transform(data_train)
  sparce_train = csr_matrix(scores_train)
  X=hstack((X, sparce_train))

  Y=vectorizer.transform(no_stop_songs_corpus)
  sparce_test = csr_matrix(scores_songs_corpus)
  Y=hstack((Y, sparce_test))

  model_gnb = MultinomialNB()

  model_gnb.fit(X, label_train)

  prediction=model_gnb.predict(Y)

  labels_songs_corpus.append(prediction)
```
